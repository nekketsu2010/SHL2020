```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.initializers import he_normal

import seaborn as sns
sns.set(font='Yu Gothic')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def user2():
    user2_Acc_x = np.loadtxt("../Data/User2/140617/Bag_Motion.txt")[:, 1].reshape([-1, 1])
    user2_Acc_x = np.concatenate([user2_Acc_x, np.loadtxt("../Data/User2/140717/Bag_Motion.txt")[:, 1].reshape([-1, 1]), np.loadtxt("../Data/User2/180717/Bag_Motion.txt")[:, 1].reshape([-1, 1])], axis=0)
    return user2_Acc_x

def user3():
    user3_Acc_x = np.loadtxt("../Data/User3/030717/Bag_Motion.txt")[:, 1].reshape([-1, 1])
    user3_Acc_x = np.concatenate([user3_Acc_x, np.loadtxt("../Data/User3/070717/Bag_Motion.txt")[:, 1].reshape([-1, 1]), np.loadtxt("../Data/User3/140617/Bag_Motion.txt")[:, 1].reshape([-1, 1])], axis=0)
    return user3_Acc_x
```

```python
user2_Acc_x = user2()
user3_Acc_x = user3()

user2_Acc_x.shape, user3_Acc_x.shape
```

```python
user2_Acc_x.shape[0] // 500, user3_Acc_x.shape[0] // 500
```

```python
user2_Acc_x = user2_Acc_x[:500*18681].reshape([-1, 500, 1])
user3_Acc_x = user3_Acc_x[:500*17536].reshape([-1, 500, 1])

user2_Acc_x.shape, user3_Acc_x.shape
```

```python
user2_Y = [0] * 18681
user3_Y = [1] * 17536
user2_Y = np.array(user2_Y).reshape([-1, 1])
user3_Y = np.array(user3_Y).reshape([-1, 1])

user2_Y.shape, user3_Y.shape
```

```python
user2_train, user2_val, user2_Y_train, user2_Y_val = train_test_split(user2_Acc_x, user2_Y, shuffle=False, test_size=0.3)
user3_train, user3_val, user3_Y_train, user3_Y_val = train_test_split(user3_Acc_x, user3_Y, shuffle=False, test_size=0.3)

user2_train.shape, user2_val.shape, user3_train.shape, user3_val.shape
```

```python
X_train = np.concatenate([user2_train, user3_train], axis=0)
Y_train = np.concatenate([user2_Y_train, user3_Y_train], axis=0)

X_val = np.concatenate([user2_val, user3_val], axis=0)
Y_val = np.concatenate([user2_Y_val, user3_Y_val], axis=0)

X_train.shape, Y_train.shape, X_val.shape, Y_val.shape
```

```python
x_mean = X_train.mean()
x_std = X_train.std()
X_train = (X_train - x_mean) / x_std
X_val = (X_val - x_mean) / x_std
```

```python
model = tf.keras.Sequential()

model.add(layers.Conv1D(128, 32, padding='valid', input_shape=(500, 1), kernel_initializer=he_normal()))
model.add(layers.Conv1D(64, 8, padding='same'))
model.add(layers.Conv1D(128, 8, padding='same'))
model.add(layers.Conv1D(32, 8, padding='same'))
model.add(layers.Conv1D(8, 8, padding='same'))
model.add(layers.Flatten())
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(2, activation='softmax'))
model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=128, batch_size=256, validation_data=(X_val, Y_val))
```
