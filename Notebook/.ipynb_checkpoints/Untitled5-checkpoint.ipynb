```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.initializers import he_normal

import seaborn as sns
sns.set(font='Yu Gothic')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def user2():
    user2_Acc_x = np.loadtxt("../Data/User2/140617/Bag_Motion.txt")[:, 1].reshape([-1, 1])
    user2_Acc_x = np.concatenate([user2_Acc_x, np.loadtxt("../Data/User2/140717/Bag_Motion.txt")[:, 1].reshape([-1, 1]), np.loadtxt("../Data/User2/180717/Bag_Motion.txt")[:, 1].reshape([-1, 1])], axis=0)
    return user2_Acc_x

def user3():
    user3_Acc_x = np.loadtxt("../Data/User3/030717/Bag_Motion.txt")[:, 1].reshape([-1, 1])
    user3_Acc_x = np.concatenate([user3_Acc_x, np.loadtxt("../Data/User3/070717/Bag_Motion.txt")[:, 1].reshape([-1, 1]), np.loadtxt("../Data/User3/140617/Bag_Motion.txt")[:, 1].reshape([-1, 1])], axis=0)
    return user3_Acc_x
```

```python
user2_Acc_x = user2()
user3_Acc_x = user3()

user2_Acc_x.shape, user3_Acc_x.shape
```

```python
val_Acc_x = np.load("../Data/センサ別npyファイル/validation/validation_Bag/validation_Bag_Acc.npy")[:14813, :, 0]
val_Acc_x.shape
```

```python
val_Acc_x = val_Acc_x.reshape([-1, 1])
val_Acc_x.shape
```

```python
np.where(val_Acc_x == user2_Acc_x)[0], np.where(user2_Acc_x == val_Acc_x)[0]
```

```python
from tqdm import tqdm
users2 = []
for i in tqdm(range(val_Acc_x.shape[0])):
    tmp = np.where(val_Acc_x[i, 0] == user2_Acc_x)[0]
    for t in tmp:
        if np.where(val_Acc_x[i, 1] == user2_Acc_x)[0][0] == (t+1):
            users2.append(t)
            break
    
```

```python
users2[0], users2[len(users2)-5]
```

```python
users2.count(2), users2.count(3), users2.count(2) + users2.count(3)
```

```python
from tqdm import tqdm
users3 = []
for i in tqdm(range(val_Acc_x.shape[0])):
    tmp = 0
    for j in range(30):
        tmp += int(np.any(user3_Acc_x == val_Acc_x[i, j]))
    if tmp >= 30:
        tmp = 3
    else:
        tmp = 2
    users3.append(tmp)
```
