```python
import numpy as np
import pandas as pd
from scipy import signal, interpolate

import tensorflow as tf

import seaborn as sns
sns.set(font='Yu Gothic')
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
def load_train(hold_position):
    file_name = "../Data/センサ別npyファイル/train/train_" + hold_position + "/train_" + hold_position
    label = np.load(file_name + "_Label.npy")
    glo_acc = np.load(file_name + "_Glo_Acc.npy")
    gyr = np.load(file_name + "_Gyr.npy")
    mag = np.load(file_name + "_Mag.npy")
    lacc = np.load(file_name + "_LAcc.npy")
    result = np.concatenate([label.copy(), glo_acc.copy(), gyr.copy(), mag.copy(), lacc.copy()], axis=2)
    del label, glo_acc, gyr, mag, lacc
    return result
```

```python
%%time
X_Bag = load_train("Bag")
X_Hips = load_train("Hips")
X_Torso = load_train("Torso")
X_Hand = load_train("Hand")
X_Bag.shape
```

## 世界標準座標系加速度Xの時間FFT

```python
nfft = 200
sensor = 1
sample_index = 50

for i in range(5):
    sample_index_tmp = sample_index + i
    label = X_Bag[sample_index_tmp, 0, 0]

    print(label)

    plt.figure(figsize=(24, 4))
    plt.subplots_adjust(wspace=0.4, hspace=0.6)

    plt.subplot(1, 4, 1)
    spectrum, freqs, t, im = plt.specgram(X_Bag[sample_index_tmp, :, sensor].copy(), NFFT=nfft, Fs=100)
    plt.grid(False)
    plt.title("Bag")
    plt.xlabel("time[second]")
    plt.ylabel("Frequnecy[Hz]")
    plt.colorbar(im)

    plt.subplot(1, 4, 2)
    spectrum, freqs, t, im = plt.specgram(X_Hips[sample_index_tmp, :, sensor].copy(), NFFT=nfft, Fs=100)
    plt.grid(False)
    plt.title("Hips")
    plt.xlabel("time[second]")
    plt.ylabel("Frequnecy[Hz]")
    plt.colorbar(im)

    plt.subplot(1, 4, 3)
    spectrum, freqs, t, im = plt.specgram(X_Torso[sample_index_tmp, :, sensor].copy(), NFFT=nfft, Fs=100)
    plt.grid(False)
    plt.title("Torso")
    plt.xlabel("time[second]")
    plt.ylabel("Frequnecy[Hz]")
    plt.colorbar(im)

    plt.subplot(1, 4, 4)
    spectrum, freqs, t, im = plt.specgram(X_Hand[sample_index_tmp, :, sensor].copy(), NFFT=nfft, Fs=100)
    plt.grid(False)
    plt.title("Hand")
    plt.xlabel("time[second]")
    plt.ylabel("Frequnecy[Hz]")
    plt.colorbar(im)
    plt.show()
```

## Scipyでスペクトログラム作った

```python
f, t, Sxx = signal.spectrogram(X_Bag[0, :, 12].copy(), fs=100)
plt.pcolormesh(t, f, Sxx)
pp=plt.colorbar (orientation="vertical") # カラーバーの表示 
pp.set_label("", fontname="Arial", fontsize=24) #カラーバーのラベル
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
def scipy_plot(sensor, sample_index):
    for i in range(5):
        sample_index_tmp = sample_index + i
        label = X_Bag[sample_index_tmp, 0, 0]

        print(label)

        plt.figure(figsize=(24, 4))
        plt.subplots_adjust(wspace=0.4, hspace=0.6)

        plt.subplot(1, 4, 1)
        f, t, Sxx = signal.spectrogram(X_Bag[sample_index, :, sensor].copy(), fs=100)
        plt.pcolormesh(t, f, Sxx)
        plt.title("Bag")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')

        plt.subplot(1, 4, 2)
        f, t, Sxx = signal.spectrogram(X_Hips[sample_index, :, sensor].copy(), fs=100)
        plt.pcolormesh(t, f, Sxx)
        plt.title("Hips")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')

        plt.subplot(1, 4, 3)
        f, t, Sxx = signal.spectrogram(X_Torso[sample_index, :, sensor].copy(), fs=100)
        plt.pcolormesh(t, f, Sxx)
        plt.title("Torso")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')

        plt.subplot(1, 4, 4)
        f, t, Sxx = signal.spectrogram(X_Hand[sample_index, :, sensor].copy(), fs=100)
        plt.pcolormesh(t, f, Sxx)
        pp=plt.colorbar (orientation="vertical") # カラーバーの表示 
        pp.set_label("", fontname="Arial", fontsize=24) #カラーバーのラベル
        plt.title("Hand")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import signal, interpolate

import tensorflow as tf

import seaborn as sns
sns.set(font='Yu Gothic')
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
def scipy_plot(sample_index):
    for i in range(5):
        sample_index_tmp = sample_index + i
        print(Label[sample_index_tmp])

        plt.figure(figsize=(24, 4))
        plt.subplots_adjust(wspace=0.4, hspace=0.6)

        plt.subplot(1, 4, 1)
        plt.pcolormesh(t, f, X_Bag[sample_index_tmp, :, :])
        plt.title("Bag")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')

        plt.subplot(1, 4, 2)
        plt.pcolormesh(t, f, X_Hips[sample_index_tmp, :, :])
        plt.title("Hips")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')

        plt.subplot(1, 4, 3)
        plt.pcolormesh(t, f, X_Torso[sample_index_tmp, :, :])
        plt.title("Torso")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')

        plt.subplot(1, 4, 4)
        plt.pcolormesh(t, f, X_Hand[sample_index_tmp, :, :])
        pp=plt.colorbar (orientation="vertical") # カラーバーの表示 
        pp.set_label("", fontname="Arial", fontsize=24) #カラーバーのラベル
        plt.title("Hand")
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.show()
```

```python
train_size = int(X_Bag.shape[0] / 2)

X_train = np.concatenate([X_Bag[:train_size], X_Hips[:train_size], X_Torso[:train_size], X_Hand[:train_size]], axis=0)
X_test = np.concatenate([X_Bag[train_size:], X_Hips[train_size:], X_Torso[train_size:], X_Hand[train_size:]], axis=0)

X_train.shape, X_test.shape
```

```python
Y_train = [0] * int(X_train.shape[0] / 4) + [1] * int(X_train.shape[0] / 4) + [2] * int(X_train.shape[0] / 4) + [3] * int(X_train.shape[0] / 4)
Y_test = [0] * int(X_test.shape[0] / 4) + [1] * int(X_test.shape[0] / 4) + [2] * int(X_test.shape[0] / 4) + [3] * int(X_test.shape[0] / 4)

Y_train = np.array(Y_train)
Y_test = np.array(Y_test)
Y_train.shape, Y_test.shape
```

```python
# ３次元のときの標準化
x_mean = X_train.mean()
x_std = X_train.std()
X_train = (X_train - x_mean) / x_std

X_test = (X_test - x_mean) / x_std
```

```python
X_train = X_train.reshape(-1, X_train.shape[1], X_train.shape[2], 1)
X_test = X_test.reshape(-1, X_test.shape[1], X_test.shape[2], 1)

X_train.shape, X_test.shape
```

```python
from tensorflow.keras import layers
from tensorflow.keras.initializers import he_normal

model = tf.keras.Sequential()

model.add(layers.Conv2D(128, (3, 3), padding='valid', activation='relu', input_shape=(X_train.shape[1], X_train.shape[2], 1), kernel_initializer=he_normal()))
model.add(layers.Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(layers.Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(layers.MaxPooling2D((2,2), padding='same'))
model.add(layers.Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(4, activation='softmax'))
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=500, batch_size=1024, validation_data=(X_test, Y_test))
```

```python
model.summary()
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.ylim((0, 2.0))
plt.show()
```
